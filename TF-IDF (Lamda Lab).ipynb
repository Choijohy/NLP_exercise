# TF-IDF Analyze


# Dataset

There are 10 sentences to create word vectors

```python
import pandas as pd
import numpy as np
```

```python
mylist = []
for i in range(1,10):
    df = pd.read_csv('Expressions%d.csv' %i ,header = None)
    list_2d= df.values.tolist()
    globals()['category'+str(i)] = np.concatenate(list_2d).tolist()
    mylist.append(np.concatenate(list_2d).tolist())
    
mylist[5]
```

```python
raw_docs = []

for i in range(9):
    raw_docs.append(" ".join(mylist[i]))
    
raw_docs
```

# Morpheme Analysis in Korean - Konlpy

```python
from konlpy.tag import Okt

okt = Okt()
corpus = []

for i in range(9):
    temp = []
    for expression in mylist[i]:
        temp+=okt.morphs(expression)
    corpus.append(temp)

                            
corpus[0]
```

# Remove Stop Words 

    * Stop words: set of commonly used words in any language
    
    In order for effiency of creating word vector, we will remove commonly used words.



```python
stopwords = ['의','가','을','이','은','들','는','좀','걍','과','도','를','으로','자','에','에만','와','한','하다','에게','.'
             ,'것','갈리','데','다','고','지','에서','']

docs = []
for i in range(9):
    for word in corpus[i]: 
        if word in stopwords:
                corpus[i].remove(word)
    docs.append(" ".join(corpus[i]))

for i in range(9):
    for word in corpus[i]:
        if len(word) == 1:
            corpus[i].remove(word)
#         docs.append(" ".join(corpus[i]))
        
for i in range(1,10):
    print("group"+str(i))
    print(docs[i-1])
    print("      ")
```

# raw_docs

```python
raw_vocab = []

for doc in docs:
    for w in doc.split(" "):
        raw_vocab.append(w)

raw_vocab = list(set(raw_vocab))
raw_vocab.sort()
```

```python
from math import log

N = len(raw_docs) 

def tf(t, d):
  return d.count(t)

def idf(t):
  df = 0
  for doc in raw_docs:
    df += t in doc
  return log(N/(df+1))

def tfidf(t, d):
  return tf(t,d)* idf(t)

result = []

# 각 문서에 대해서 아래 연산을 반복
for i in range(N):
  result.append([])
  d = raw_docs[i]
  for j in range(len(raw_vocab)):
    t = raw_vocab[j]
    result[-1].append(tf(t, d))

tf_ = pd.DataFrame(result, columns = raw_vocab)

tf_
```

```python
result = []
for j in range(len(raw_vocab)):
    t = raw_vocab[j]
    result.append(idf(t))

idf_ = pd.DataFrame(result, index=raw_vocab, columns=["IDF"])
idf_
```

```python
result2 = []
for i in range(N):
  result2.append([])
  d = raw_docs[i]
  for j in range(len(raw_vocab)):
    t = raw_vocab[j]
    result2[-1].append(tfidf(t,d))

tfidf_ = pd.DataFrame(result2, columns = raw_vocab)
tfidf_
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
vect = CountVectorizer()
document_term_matrix = vect.fit_transform(docs)       # 문서-단어 행렬 

tf = pd.DataFrame(document_term_matrix.toarray(), columns=vect.get_feature_names_out())
                                             # TF (Term Frequency)
D = len(tf)
df = tf.astype(bool).sum(axis=0)
idf = np.log((D+1) / (df+1)) + 1             # IDF (Inverse Document Frequency)

# TF-IDF (Term Frequency-Inverse Document Frequency)
tfidf = tf * idf                      
tfidf = tfidf / np.linalg.norm(tfidf, axis=1, keepdims=True)

tfidf =round(tfidf, 3)

tfidf
```

```python
idf_df = pd.DataFrame(idf)
idf_df=idf_df.sort_values(by=0, ascending=False)
```
