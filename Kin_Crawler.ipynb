```python
import pandas as pd
import numpy as np
import platform
import matplotlib.pyplot as plt

from bs4 import BeautifulSoup 
from urllib.request import urlopen
import urllib
import time
```

```python
#지식in url
#1 페이지
url = "https://kin.naver.com/search/list.naver?query=%EC%9A%B0%EC%9A%B8%EC%A6%9D+%ED%94%BC%EB%A1%9C"
url_list = []

page_url = url
html = urlopen(page_url)
soup = BeautifulSoup(html,"html.parser",from_encoding='utf-8')
articles = soup.findAll("dt")

for link in articles:
    url_list.append(link.find("a")["href"])

#2페이지 ~ 100
num = 25
# 100페이지까지
for i in range(2,num):
    #지식in url 
    url = "https://kin.naver.com/search/list.naver?query=%EC%9A%B0%EC%9A%B8%EC%A6%9D+%ED%94%BC%EB%A1%9C&page="+str(i)

    page_url = url
    html = urlopen(page_url)
    soup = BeautifulSoup(html,"html.parser",from_encoding='utf-8')
    articles = soup.findAll("dt")

    for link in articles:
        url_list.append(link.find("a")["href"])

```

```python
import re 

texts = []
titles = []
answers = []


for url in url_list2:
    html_article = urlopen(url.encode('ascii', 'ignore').decode('ascii'))
    soup_article = BeautifulSoup(html_article,"html.parser",from_encoding='utf-8')

    if(soup_article.find("div",{"class":"c-heading__content"}) == None):
        content = soup_article.find("div",{"class":"title"}).text
        title = ""
    else:  
        content = soup_article.find("div",{"class":"c-heading__content"}).text
        title = soup_article.find("div",{"class":"title"}).text

    #     answer_test =  soup_article.find("div",{"class":"_endContentsText"}).text


    p = re.compile('\<.*?\>') #괄호안에 있는거 다 지우겠다.

    test = str(content)
    title_test = str(title)
    #     answer_test = str(answer_test)

    real_text = re.sub(p, "", test)
    real_text = re.sub("\t", "", real_text)
    real_text = re.sub("\n", "", real_text)

    title = re.sub(p, "", title_test)
    title = re.sub("\t", "", title)
    title = re.sub("\n", "", title)

    #     answer = re.sub(p, "", answer_test)
    #     answer = re.sub("\t", "", answer)
    #     answer = re.sub("\n", "", answer)


    titles.append(title)
    texts.append(real_text)
    #     answers.append(answer)


# data = [titles, texts]

# newData = np.transpose(data)

# df = pd.DataFrame(newData, columns = ["제목","질문"])

# # df.to_csv("C:\\Users\\gunco\\Desktop\\df.csv")

# df
```
