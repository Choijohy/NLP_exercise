```python
import pandas as pd

df = pd.read_csv('kin_depression.csv')
```

```python
import re
import urllib.request
import zipfile
from lxml import etree
from konlpy.tag import Okt
```

```python
df=df['질문'].values.tolist()
```

```python
print(len(df))
```

```python
print(df[:5]) # df=['sentence1','sentence2'...]
```

```python
normalized_text = []
for string in df: #한글만 추출  : 특수문자, 숫자, 영어 모두 제거
    hangul = re.compile('[^ ㄱ-ㅣ가-힣+]') 
    result = hangul.sub('', string)
    normalized_text.append(result)
print(normalized_text[:5])
```

```python
# remove stopwords
okt = Okt()
corpus = []

for i in normalized_text:
    corpus.append(okt.morphs(i))
corpus
```

```python
len(corpus)  # corpus=[['token1','token2'...],['token1','token2']...] : 1018개의 토큰화된 문장
```

```python
stop_word_list = pd.read_csv('stopword.txt',sep = "\t")
stop_word_list = stop_word_list.values.tolist()
```

```python
stop_words = []
for l in stop_word_list:
    for w in l: 
        stop_words.append(w)
```

```python
print("중복 제거 전: ",len(stop_words))
stop_words = list(set(stop_words))
print("중복 제거 후: ",len(stop_words))
```

```python
def remove_stop_words(corpus,stop_words):
    cleaned_corpus = []

    print(stop_words[:10])
    for s in corpus: 
        result = []
        for token in s: 
            if token not in stop_words: 
                result.append(token)
        cleaned_corpus.append(result)             
                
    return cleaned_corpus

result_data = remove_stop_words(corpus,stop_words)
```

```python
result_data[5:10]
```

```python
print('총 샘플의 개수 : {}'.format(len(result_data)))
```

```python
# top 3 samples for W2V
for line in result_data[0:3]:
    print(line)
```

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors

model = Word2Vec(sentences=result_data,window=10, min_count=5, workers=4, sg=0)
```

```python
model_result = model.wv.most_similar("친구")
print(model_result)
```

```python
model.wv.save_word2vec_format('kin_w2v', binary=False)
```
